```python
# Third-party modules
import seaborn as sns
import matplotlib.pyplot as plt

# Custom modules
from src.entities.ppi import PPI
from src.misc.logger import logger
```

```python
ppis = [ppi for ppi in PPI.iterate('distance_map.CA') if ppi.interact != '?']
```

```python
# Map size as much as seq1 + seq2 (no C-domain removed)
map_sizes = [len(ppi.distance_map['CA']) for ppi in ppis]
logger.info(f'Min: {min(map_sizes)}')
logger.info(f'Max: {max(map_sizes)}')
logger.info(f'Mean: {sum(map_sizes) / len(map_sizes)}')
```

```python
# Plot density plot
plt.figure(figsize=(10, 6))
sns.histplot(map_sizes, bins=30, kde=True, color='blue', stat='density')
plt.title('Distribution of Map Sizes')
plt.xlabel('Map Size')
plt.ylabel('Density')
```
